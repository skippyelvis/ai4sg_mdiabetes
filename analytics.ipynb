```python
import torch
import matplotlib.pyplot as plt

def load_debug(f, w):
    return torch.load(f'local_storage/{f}/debug/{w}.pt')
    
def plot_loss(f, w):
    d = load_debug(f, w)
    plt.plot(d['loss'])
    
def plot_metrics(f, w_max):
    h = []
    s = []
    u = []
    fig, ax = plt.subplots(ncols=2)
    for w in range(w_max):
        db = load_debug(f, w+1)
        try:
            s.append(db['metrics']['average_score'])
            u.append(db['metrics']['uniqueness'])
            h.append(w)
        except:
            pass
    ax[0].bar(h, s)
    ax[1].bar(h, u)
    ax[0].set_ylim(0,8)
    ax[1].set_ylim(0,1)
```

```python
plot_metrics('regularized_testing_9', 14)
```

```python
plot_loss('regularized_testing_2', 8)
```
